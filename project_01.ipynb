```python
import pandas as pd 

from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error , mean_squared_error 

import numpy as np 

data = pd.read_csv(r"Datasets\Test Score - Sheet1.csv")


X = data [['Hours']]

y = data['Score']

model = LinearRegression()

model.fit(X,y)

predected_score = model.predict(X)


#evaluation

mae = mean_absolute_error(y,predected_score)
mse = mean_squared_error(y, predected_score)
rmse = np.sqrt(mse)

print("Mean absolute error :",mae)
print("Mean squared error :",mse)
print("root mean squared error :",rmse)

new_hour = float(input("enter an hour"))
new_pred = model.predict([[new_hour]])
print(f"Prediction for {new_hour}is score = {new_pred}")
```
