```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


# Sample data 

data = {
    'Customer':['Riya','Aman','Faizan','Harsh','puja','Saubhagya','keshav'],
    'Age':[20,30,40,22,38,25,45],
    'Spending':[100,200,300,110,290,130,200]
}

df = pd.DataFrame(data)

X = df[['Age','Spending']]

model = KMeans(n_clusters=2,random_state=42,n_init=10)

df['Group'] = model.fit_predict(X)

plt.figure(figsize=(6,5))
for Group in df['Group'].unique():
    group_data = df[df['Group']== Group]
    plt.scatter(group_data['Age'],group_data['Spending'],label=f'Group{Group}')


plt.xlabel('Age')
plt.ylabel('Spending')
plt.title('Customer segements(k-means)')
plt.legend()
plt.grid(True)
plt.show()

print(df)
```
